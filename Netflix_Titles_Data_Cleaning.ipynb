DATA UNDERSTANDING

```python
import pandas as pd

#Loading the dataset
df = pd.read_csv("netflix_titles.csv")

```

```python
#qn1: Rows and columns in the dataset
rows,cols = df.shape
print(f"The dataset contains {rows} rows and {cols} columns. \n")
```

```python
#qn2: Datatypes contained in the datasets
print("Data types of each column:")
print(df.dtypes, "\n")

```

```python
#qn3: Is the datasets complete?
missing_values = df.isnull().sum()
total_missing = missing_values.sum()
if total_missing == 0:
  print("The dataset is complete, no missing values found.\n")
else:
  print("The dataset is not complete, there are missing values in some columns.\n")
```

```python
#qn4: The number of moviesand TV shows in the last 20 rows
last_20 = df.tail(20)
movie_count = last_20[last_20['type'] == 'Movie'].shape[0]
tvshow_count = last_20[last_20['type'] == 'TV Show'].shape[0]
print(f"Among the last 20 rows, there are {movie_count} movies and {tvshow_count} TV shows.")
```

DATA INDEXING

```python
#QN1: Movies originating from South Africa
south_africa_movies = df[(df['type'] == 'Movie') & (df['country'] == 'South Africa')]
num_south_africa_movies = south_africa_movies.shape[0]
print(f"Number of movies from South Africa are {num_south_africa_movies}\n")
```

```python
#QN2: Movie in the 12th row
description_12th = df.iloc[11]['description']
print(f"Description of the 12th movie: \n")
description_12th
```

```python
#QN3: South African movies released in 2021
south_africa_2021 = south_africa_movies[south_africa_movies['release_year'] == 2021]
print("South African movies released in 2021:\n")
south_africa_2021[['title', 'release_year']]
```

```python
#QN4: Entire cast of all movies released in 2021
movies_2021 = df[(df['type'] == 'Movie') & (df['release_year'] == 2021)]
casts_2021 = movies_2021[['title', 'cast']]
print("Cast of all movies released in 2021:")
casts_2021
```
